```python
import os
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from tensorflow.keras.metrics import Precision,Recall,F1Score
```

```python
class Cherry():

  def __init__(self):

    self.data_dir='/content/drive/MyDrive/Cherry/train' #trainig data directory in google drive
    self.data_dir_test='/content/drive/MyDrive/Cherry/val'  #testing data directory in google drive
    self.train_size, self.test_size=self.data_take() #obtaining train and test size based on length of dataset
    self.prec=Precision() #evaluation metrics
    self.recall=Recall() #evaluation metrics
    self.f1=F1Score()  #evaluation metrics
    self.saved_model=load_model('/content/drive/MyDrive/Cherry/cherry_model.h5')  #model saved after trainig to predict unseen images


  def data_take(self):

    self.data=tf.keras.utils.image_dataset_from_directory(self.data_dir) #generates a Dataset of images from its directory
    return int(len(self.data)*0.7), int(len(self.data)*0.3) #output train and test length


  def pre_processing_data(self):

    self.data=self.data.map(lambda x,y:(x/255,y)) #preprocessing


  def train_test_split(self):

    return self.data.take(self.train_size), self.data.skip(self.train_size).take(self.test_size) #spliting dataset to train and test


  def get_test_data(self):

    self.val_data=tf.keras.utils.image_dataset_from_directory(self.data_dir_test) #generates test data set from directory
    self.val_data=self.val_data.map(lambda x,y:(x/255,y)) #preprocessing
    self.val_data=self.val_data.take(len(self.val_data))

    return self.val_data



  def cnn_model_arch(self):

    self.model=Sequential() #CNN model with maxpooling and dropout layers

    self.model.add(Conv2D(64,(3,3),1, activation='relu', input_shape=(256,256,3)))
    self.model.add(MaxPooling2D())

    self.model.add(Conv2D(128,(3,3),1, activation='relu', input_shape=(256,256,3)))
    self.model.add(MaxPooling2D())
    self.model.add(Dropout(0.5))

    self.model.add(Conv2D(64,(3,3),1, activation='relu', input_shape=(256,256,3)))
    self.model.add(MaxPooling2D())
    self.model.add(Dropout(0.3))

    self.model.add(Conv2D(32,(3,3),1, activation='relu', input_shape=(256,256,3)))
    self.model.add(MaxPooling2D())
    self.model.add(Dropout(0.6))

    self.model.add(Conv2D(64,(3,3),1, activation='relu', input_shape=(256,256,3)))
    self.model.add(MaxPooling2D())
    self.model.add(Dropout(0.2))

    self.model.add(Conv2D(32,(3,3),1, activation='relu', input_shape=(256,256,3)))
    self.model.add(MaxPooling2D())

    self.model.add(Flatten())

    self.model.add(Dense(128,activation='relu'))
    self.model.add(Dense(1,activation='sigmoid'))

    self.model.compile(optimizer=tf.keras.optimizers.Adam(),
                loss=tf.keras.losses.BinaryCrossentropy(),
                  metrics=['accuracy'])
    return self.model

  def train_model(self):

    self.model=self.cnn_model_arch()
    logdir='logs'
    tensorboard_callback=tf.keras.callbacks.TensorBoard(log_dir=logdir)
    train, test=self.train_test_split()
    self.hist=self.model.fit(train,epochs=4,
                   callbacks=[tensorboard_callback],
                   validation_data=test)  

    self.model.save('/content/drive/MyDrive/Cherry/cherry_model.h5')  # trained model saved


  def plot_loss_accuracy(self): #plot of loss and accuracy

    fig,ax=plt.subplots(2,figsize=(10,10))

    ax[0].plot(self.hist.history['loss'],label='loss')
    ax[0].plot(self.hist.history['val_loss'],label='val loss')
    ax[0].legend()

    ax[1].plot(self.hist.history['accuracy'],label='accuracy')
    ax[1].plot(self.hist.history['val_accuracy'],label='val accuracy')
    ax[1].legend()

    plt.tight_layout()

    plt.show()


  def test_model(self): #testing with new datset

    self.val_data=self.get_test_data()
    i=0

    for batch in self.val_data.as_numpy_iterator():
      X, y=batch
      y_pred=self.saved_model.predict(X)
      self.prec.update_state(y, y_pred)
      self.recall.update_state(y, y_pred)
      # self.f1.update_state(y, y_pred)
      # print('batch', i)
      # i+=1

    print(f'Precision:{self.prec.result().numpy()}, Recall:{self.recall.result().numpy()}')

```

```python
c=Cherry()
c.pre_processing_data()
c.train_model()
```

```python
c.plot_loss_accuracy()
```

```python
c=Cherry()
c.test_model()
```
